```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
DATA_DIR = 'data'
train_df = pd.read_csv(os.path.join(DATA_DIR, 'data/train.csv'))
```

```python
train_df.head()
```

```python
train_df['Gender'].unique()
```

```python
train_df.isna()
```

```python
na = []
for col in train_df.columns:
  na.append(sum(train_df[col].isna()))
print(na)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(train_df['Gender'])
train_df['Gender'] = label_encoder.fit_transform(train_df['Gender'])
train_df.head()
```

```python
label_encoder.get_params()
```

```python
keys = label_encoder.classes_
values = label_encoder.transform(keys)
labels = dict(zip(keys, values))
labels
```

```python
train_x = train_df.drop(['id', 'Target'], axis=1)
train_y = train_df['Target']
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(train_x, train_y)
```

```python
test_df = pd.read_csv(os.path.join(DATA_DIR, 'data/test.csv'))
test_df.head()
```

```python
test_df['Gender'] = label_encoder.fit_transform(test_df['Gender'])
test_df.head()
```

```python
test_x = test_df.drop(['id'], axis=1)
```

```python
prediction = model.predict(test_x)
```

```python
prediction = np.round(prediction)
```

```python
submission = pd.read_csv(os.path.join(DATA_DIR, 'data/sample_submission.csv'))
submission['Target'] = prediction
```

```python
submission.to_csv('submit.csv', index=False)
```

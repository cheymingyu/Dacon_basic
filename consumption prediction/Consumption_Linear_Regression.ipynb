```python
import pandas as pd
import numpy as np
import os
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
submission = pd.read_csv('sample_submission.csv')
```

```python
train.head()
```

```python
# Dt_Customer를 연, 월, 일으로 분리
def seperate_datetime(dataframe):
  year = []
  month = []
  day = []

  for date in dataframe['Dt_Customer']:
    day_point, month_point, year_point = date.split('-')
    year.append(int(year_point))
    month.append(int(month_point))
    day.append(int(day_point))
  return year, month, day

year, month, day = seperate_datetime(train)
train['year'] = year
train['month'] = month
train['day'] = day

train.head()
```

```python
train.drop('Dt_Customer', axis=1, inplace=True)
train.head()
```

```python
# test 데이터도 똑같이 진행
year, month, day = seperate_datetime(test)
test['year'] = year
test['month'] = month
test['day'] = day
test.drop('Dt_Customer', axis=1, inplace=True)
test.head()
```

```python
# categorical: Education, Marital_Status
# Label Encoding 진행
def make_label_map(dataframe):
  label_maps = {}
  for col in dataframe.columns:
    if dataframe[col].dtype == 'object':
        label_map = {'unknown': 0}  # 혹시 모를 결측값에 대비해 unknown은 0으로 채운다.
        for i, key in enumerate(dataframe[col].unique()):
          label_map[key] = i
        label_maps[col] = label_map
  return label_maps

def label_encoding(dataframe, label_map):
  for col in dataframe.columns:
    if dataframe[col].dtype == 'object':
      dataframe[col] = dataframe[col].map(label_map[col])
  return dataframe

label_map = make_label_map(train)
train = label_encoding(train, label_map)
train.head()
```

```python
# test에도 적용
test = label_encoding(test, label_map)
test.head()
```

```python
from sklearn.model_selection import train_test_split
train_data, val_data = train_test_split(train, test_size=0.3)
train_data.reset_index(inplace=True)
val_data.reset_index(inplace=True)
```

```python
train_X = train_data.drop('target', axis=1)
train_y = train_data['target']
val_X = val_data.drop('target', axis=1)
val_y = val_data['target']
```

```python
train_X.shape, val_X.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(train_X, train_y)
```

```python
from sklearn.metrics import mean_squared_error

def nmae(true, pred):
  mae = np.mean(np.abs(true - pred))
  score = mae / np.mean(np.abs(true))

  return score

y_hat = model.predict(val_X)
print(f'모델 NMAE: , {nmae(val_y, y_hat)}')
```

```python
X = train.drop('target', axis=1)
y = train['target']

model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(test)
```

```python
submission['target'] = y_pred
submission.to_csv('submission.csv', index=False)
```
